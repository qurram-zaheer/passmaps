```python
import matplotlib.pyplot as plt
from matplotlib.patches import Arc
import pandas as pd
import seaborn as sns
```

```python
def pitch():
    fig=plt.figure(figsize=(13,9))
    ax=fig.add_subplot(1,1,1)
    ax.set_facecolor('#22312b')
    fig.patch.set_facecolor('#22312b')
    ax.xaxis.label.set_color('white')        
    ax.yaxis.label.set_color('white')          

    # ax.spines['left'].set_color('white') 
    # ax.spines['top'].set_color('white')
    # ax.spines['bottom'].set_color('white')
    # ax.spines['right'].set_color('white') 
    plt.axis('off')


    ### Pitch outline
    plt.plot([0,0], [0,90], color='white')
    plt.plot([0,130],[90,90], color='white')
    plt.plot([65,65],[0,90], color='white')
    plt.plot([130,130],[0,90], color='white')
    plt.plot([130,0],[0,0], color='white')

    ### Left Penalty Area
    plt.plot([16.5,16.5],[65,25], color='white')
    plt.plot([0,16.5],[65,65], color='white')
    plt.plot([16.5,0],[25,25], color='white')

    leftArc=Arc((11,45),height=18.3,width=18.3,angle=0,theta1=310,theta2=50,color="white")

    #Right Penalty Area
    plt.plot([130,113.5],[65,65],color="white")
    plt.plot([113.5,113.5],[65,25],color="white")
    plt.plot([113.5,130],[25,25],color="white")

    #Left 6-yard Box
    plt.plot([0,5.5],[54,54],color="white")
    plt.plot([5.5,5.5],[54,36],color="white")
    plt.plot([5.5,0],[36,36],color="white")
    plt.plot([0,0],[41.34,48.66], color='white',linewidth=4)

    #Right 6-yard Box
    plt.plot([130,124.5],[54,54],color="white")
    plt.plot([124.5,124.5],[54,36],color="white")
    plt.plot([124.5,130],[36,36],color="white")
    plt.plot([130,130],[41.34,48.66], color='white',linewidth=4)

    #Prepare Circles
    centreCircle = plt.Circle((65,45),9.15,color="white",fill=False)
    centreSpot = plt.Circle((65,45),0.8,color="white")
    leftPenSpot = plt.Circle((11,45),0.8,color="white")
    rightPenSpot = plt.Circle((119,45),0.8,color="white")

    #Draw Circles
    ax.add_patch(centreCircle)
    ax.add_patch(centreSpot)
    ax.add_patch(leftPenSpot)
    ax.add_patch(rightPenSpot)

    #Prepare Arcs
    leftArc = Arc((11,45),height=18.3,width=18.3,angle=0,theta1=308,theta2=52,color="white")
    rightArc = Arc((119,45),height=18.3,width=18.3,angle=0,theta1=128,theta2=232,color="white")

    #Draw Arcs
    ax.add_patch(leftArc)
    ax.add_patch(rightArc)

    
    return ax, fig
```

```python
df=pd.read_csv('map.csv')
df[['x','endX']]=df[['x','endX']]*1.3
df[['y','endY']]=df[['y','endY']]*0.9
```

```python
df
```

```python
ax, fig=pitch()

kde=sns.kdeplot(
    df['x'],
    df['y'],
    clip=((0,130),(0,90)),
    shade=True,
    shade_lowest=False,
    alpha=0.2,
    n_levels=10,
    cmap='magma'
)

for index,row in df.iterrows():
    if row['outcome'] == 'Successful':
        plt.plot([row['x'],row['endX']],[row['y'],row['endY']], color='#42f578', alpha=0.5)
        plt.scatter(row['x'], row['y'], color='#42f578', alpha=1)
    else:
        plt.plot([row['x'],row['endX']],[row['y'],row['endY']], color='#f5425a', alpha=0.5)
        plt.scatter(row['x'], row['y'], color='#f5425a', alpha=0.5)
plt.xlim=(0,130)
plt.ylim=(0,90)
plt.show()
```
